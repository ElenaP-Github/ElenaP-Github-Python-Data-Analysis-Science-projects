# Some basic visualizations

## Histograms

Use: summarize discrete or continous data

The height of the bar shows the number of times that the values occurred within the interval, while the width of the bar shows the interval that is covered. For a histogram with equal bins, the width should be the same across all bars. [[reference](https://corporatefinanceinstitute.com/resources/excel/study/histogram/)]

```python
fig, ax = plt.subplots(figsize=(20,5))
df1['A'].plot.kde(ax=ax, legend=False, title='Histogram: column B', c="red", lw=3);## density line
df1['A'].plot.hist(ax=ax,density=True, bins=20,edgecolor='k', grid=True).autoscale(enable=True, axis= "both", tight=True );
```

```python
# df2.plot.kde();### density plot
fig, ax = plt.subplots(figsize=(20,5))
df2[["b"]].plot.kde(ax=ax, legend=False, title='Histogram: column c', lw=3);## density line
df2[["b"]].plot.hist(ax=ax,density=True, bins=20,edgecolor='k', grid=True, stacked=False).autoscale(enable=True, axis= "both", tight=True );
```

## Bar plot

A bar chart **plots numeric values for levels of a categorical feature as bars**. Levels are plotted on one chart axis, and values are plotted on the other axis. 

Each categorical value claims one bar, and the length of each bar corresponds to the bar’s value. Bars are plotted on a common baseline to allow for easy comparison of values.

[[reference](https://chartio.com/learn/charts/bar-chart-complete-guide/)]

```python
## Bar charts using the date as index

ax=df1[0:8].plot.bar(stacked=True);## vertical direction 
ax.legend(loc=0, bbox_to_anchor=(1.2,1.0));
```

```python
## bar charts without index

ax=df2.plot.barh(stacked=True);
ax.legend(loc=0, bbox_to_anchor=(1.2,1.0));#·horizontal direction
```

## Line chart

Just for continous data. It is useful when one is trying to identify a trend or pattern. 


```python
ax= df2.plot.line(y=["a", "b", "c"],figsize=(10,4), lw=2, ls="--", marker="o", grid=True);
ax.set(title="My line plot ", xlabel="My x data",ylabel="My y data");
ax.legend(loc=0, bbox_to_anchor=(1.0,1.0));## legend outside of the graph
```

## Area chart

An area chart is really similar to a line chart, except that the area between the x axis and the line is filled in with color or shading. It represents the evolution of a numerical variable following another numerical variable
[[reference]( https://python-graph-gallery.com/area-plot/)]

```python
ax= df2.plot.area(stacked=False);
ax.legend(loc=0, bbox_to_anchor=(1.0,1.0));## legend outside of the graph
```

## Scatter plot

A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

Identification of correlational relationships are common with scatter plots

[[reference](https://chartio.com/learn/charts/what-is-a-scatter-plot/#when-you-should-use-a-scatter-plot)]

```python
df1.plot.scatter(x="A", y="D", c="C", cmap="coolwarm", alpha =0.5);
```

```python
import numpy as np 
df=pd.DataFrame(np.random.randn(1000,2), columns=["a","b"])### 1000 rows and two columns
df.plot.hexbin(x="a",y="b", gridsize=25, cmap="coolwarm");
```

## Box plot

Box plots divide the data into sections that each contain approximately 25% of the data in that set. Box plots are useful as they provide a visual summary of the data enabling researchers to quickly identify mean values, the dispersion of the data set, and signs of skewness.

**Note**: 

The median is generally considered to be the best representative of the central data values than the mean. The more skewed the distribution, the greater the difference between the median and mean. 

The mean is much more sensitive to extreme values than the median. That is, one or two extreme values can change the mean a lot but do not change the the median very much.



[[reference](https://www.simplypsychology.org/boxplots.html)]

```python
df2.plot.box(figsize=(8,5));
```

```python
df2.boxplot(grid=False, figsize=(8,5));
```
